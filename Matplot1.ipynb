```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
a=pd.read_csv("gas_prices.csv",keep_default_na=False)
a
```

```python
x=[1,2,3]
y=[2,4,6]
plt.plot(x,y,label="2x",linestyle="--",marker=".",linewidth=2,color="red",markeredgecolor="blue")
plt.title("First Graph",fontdict={"fontname":"Comic Sans Ms","fontsize":20})
x2=np.arange(0,4.5,0.5)
plt.plot(x2[:6],x2[:6]**2,"r",label="X^2")
plt.plot(x2[5:],x2[5:]**2,"r--")
plt.xlabel("X axis")
plt.ylabel("Y Axis")
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(8,5))
plt.plot(a.Year,a.Germany,"r.-",label="Germany")
plt.plot(a.Year,a.USA,"b.-",label="USA")
plt.title("Gas Prices in USD",fontdict={"fontname":"Comic Sans Ms","fontsize":20})
plt.xlabel("Years")
plt.ylabel("Prices In USD")
plt.xticks(a.Year[::3])
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(8,5))
for country in a:
    if country!="Year":
        plt.plot(a.Year,a[country],marker=".",label=country)
        
plt.xticks(a.Year[::3])       
plt.title("Gas Prices in USD",fontdict={"fontname":"Comic Sans Ms","fontsize":20})
plt.xlabel("Years")
plt.ylabel("Prices In USD")
plt.legend()
plt.show()
```

```python
#plt.plot(a.Year,a.Australia,label="Australia")
plt.plot(a.Year,a.Canada,label="Canada")
plt.plot(a.Year,a.Australia,label="Australia")
plt.legend()
plt.xlabel("Years")
plt.ylabel("Prices in USD")
plt.show()
```

```python
plt.figure(figsize=(8,5))
countrys_to_look=["USA","Australia","Canada","South Korea"]
for country in a:
    if country in countrys_to_look:
        plt.plot(a.Year,a[country],label=country,marker=".")
plt.xticks(a.Year[::3])       
plt.title("Gas Prices in USD",fontdict={"fontname":"Comic Sans Ms","fontsize":20})
plt.xlabel("Years")
plt.ylabel("Prices In USD")
plt.legend()
plt.show()
```
